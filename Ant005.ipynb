```python
dataset['year'] = pd.DatetimeIndex(dataset['timestamp']).year

dataset['year'].value_counts()
```

```python
dataset['month'] = pd.DatetimeIndex(dataset['timestamp']).month

dataset['month'].value_counts()
```

```python
dataset.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
category_counts=dataset['category'].value_counts()

plt.bar(category_counts.index,category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of items in each category')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.hist(dataset['rating'],bins=10)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of ratings')
plt.show()
```

```python
model_attr_counts = dataset['model_attr'].value_counts()

plt.pie(model_attr_counts, labels=model_attr_counts.index,autopct='%1.1f%%', startangle=90)
plt.title('Distribution of model_arr')
plt.show()
```

```python
dataset['timestamp'] = pd.to_datetime(dataset['timestamp'])
dataset['month'] = dataset['timestamp'].dt.month

monthly_avg_rating = dataset.groupby('month')['rating'].mean()

plt.plot(monthly_avg_rating.index, monthly_avg_rating.values)
plt.xlabel('Month')
plt.ylabel('Average Rating')
plt.title('Average Rating Over Time')
plt.grid(True)
plt.show()
```

```python
plt.scatter(dataset['year'], dataset['rating'])
plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Rating vs Year')
plt.grid(True)
plt.show()

```
